```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x="salary",y="sales",data=df,hue="level of education"
,palette="Dark2",size="salary",s=200,style="level of education")
#alpha=0.
plt.savefig("my_seaborn_plot.jpg")
```

```python
plt.figure(figsize=(5,8),dpi=200)
sns.rugplot(x="salary",data=df,height=0.5)

```

```python
sns.set(style="whitegrid")
sns.displot(data=df,x="salary",bins=10,color="red",edgecolor="blue"
,linewidth=4,ls="--",kde=True,rug=True )
```

```python
sns.kdeplot(data=df,x="salary")
```

```python
np.random.seed(42)
sample_ages=np.random.randint(0,100,200)
sample_ages=pd.DataFrame(sample_ages,columns=['age'])
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages,height=0.5)
```

```python
sns.displot(data=sample_ages,x="ages",rug=True,bins=30,kde=True)
```

```python
sns.kdeplot(data=sample_ages,x="ages",clip=[0,100],bw_adjust=0.01  )
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='division')
```

```python
df['division'].value_counts()
```

```python
sns.countplot(data=df,x='level of education',hue="division",palette='Set2')
```

```python
sns.barplot(data=df,x="level of education",y="salary",estimator=np.mean,ci="sd")
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=df,x="level of education",y="salary",estimator=np.mean,ci="sd",hue="division")
plt.legend(bbox_to_anchor=(1.05,1))
```
