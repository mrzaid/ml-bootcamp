```python
def speeding_game(speed,birthday):

    if birthday:
        speeding=speed - 5
    else:
        speeding=speed
    if speeding >80:
        return "Big Ticket"
    elif speeding >60:
        return "Small Ticket"
    else:
        return "No Ticket"
speeding_game(85,True)
```
